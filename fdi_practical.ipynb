```python
from tqdm.notebook import trange, tqdm

def stupid_ex(base, exponent, modulus):
    result = base
    for i in trange(1, exponent):
        result = result*base

    return result % modulus

def better_ex(base, exponent, modulus):
    result = base
    for i in trange(1, exponent):
        result = result*base
        result = result % modulus
        
    return result

base = 252
exponent = 473
modulus = 133

stupid_remainder = stupid_ex(base, exponent, modulus)
better_remainder = better_ex(base, exponent, modulus)
print(stupid_remainder, better_remainder)
```

```python
from tqdm.notebook import trange, tqdm


def sqm(base, exponent, modulus):
    result = base
    binary_exponent = [ b for b in bin(exponent)[3:]]

    for coefficient in tqdm(binary_exponent):
        result = result * result
        print(coefficient, result)
        if (coefficient == '1'):
            result = result * base
        result = result % modulus

    return result

a = 252
b = 10**600
sqm(a, b, 1000)
```
